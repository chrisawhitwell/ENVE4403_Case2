### Example: Velocity profile in a pipe
Sometimes we non-dimensionalise our axes so that we can generalise the output. Consider the velocity profile in a horizontal laminar pipe flow with radius $R$ driven by a pressure gradient $\alpha$ for a fluid with dynamic viscosity $\mu$. The velocity profile as a function of radial distance $r$ from the center , $u(r)$  is given by:
$$
u(r) = \alpha\frac{R^2 - r^2}{4\mu}
$$

Say that we want to plot this function with the radial distance on the y-axis and the velocity on the x axis but want to make it is generic as possible, we can non-dimensionalise y-axis by dividing by $R$ and the x-axis by the maximum velocity $u_{max}$, such that the x axis is 1 at the maximum velocity, and the y axis is 0 at the center of the pipe, and +-1 at the edges of the pipe.

The maximum velocity occurs at the center of the pipe, $r=0$, and is given by:
$$
u_{max} = \alpha\frac{R^2}{4\mu}
$$

So we can write the velocity profile as a function of $r/R$ and $u/u_{max}$ as:
$$
u/u_{max} = 1 - (\frac{r}{R})^2
$$


```python
r_R = np.linspace(-1,1,100)
u_umax = 1- r_R**2

fig,axes = plt.subplots()
axes.plot(u_umax,r_R,label='Profile')
axes.set_xlabel('u/umax')
axes.set_ylabel('r/R')

axes.axhspan(1,1.1,facecolor='0.5',alpha=0.5,label='wall')
axes.axhspan(-1.1,-1,facecolor='0.5',alpha=0.5)
axes.legend(loc='upper right')
axes.set_ylim(-1.1,1.1)
```

Note that there are no units for dimensionless axes.
